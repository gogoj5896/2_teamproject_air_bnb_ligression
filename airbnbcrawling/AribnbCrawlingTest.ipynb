# 크롤링

```python
%time # 이 코드를 돌리는데 얼마나 시간이 걸리는지 측정
for i in range(9000, 9999): # 몇 번째 room_id를 크롤링할 것인지 결정
    rid = df[df.city == 'Switzerland']['room_id'].iloc[i] # 도시명
    if i % 10 == 0: print((rid,i), '*' * 40) # 10개마다 출력
    crawler(str(rid), dic=test, prt=False)
    time.sleep(0.2) # 크롤링 대기시간
df_test = pd.DataFrame(test)
df_test.columns.name = 'room_id'
df_test.index.name = 'features'
df_test = df_test.stack().unstack('features') # row와 column 바꿈
df_test.to_csv('~/hosthome/1_data/data_new.csv', sep=',', encoding='utf-8') # 지정된 이름의 csv파일로 저장
```
